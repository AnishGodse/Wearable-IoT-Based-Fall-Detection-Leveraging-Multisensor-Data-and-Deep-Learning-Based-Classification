```python
import pandas as pd
import numpy as np
from scipy.stats import entropy, iqr, kurtosis, skew, pointbiserialr
from scipy.signal import welch
from scipy.fftpack import fft
from sklearn.model_selection import train_test_split
from sklearn.utils import class_weight
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers, models, callbacks
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# === 1. Load and preprocess training data ===
# Use the training dataset (update the file path as needed)
df = pd.read_csv("/content/final_bpm_input_train.csv")

# Do NOT drop bpm now, since we want to include it as a sensor feature
# if "bpm" in df.columns:
#     df = df.drop(columns=["bpm"])
df['label'] = df['label'].fillna(0)

# === 2. Create batches of 42 rows and extract features ===
# Update sensor_columns to include bpm
sensor_columns = ["AcX", "AcY", "AcZ", "GyX", "GyY", "GyZ", "bpm"]

batch_size = 42
num_batches = len(df) // batch_size
features_list = []
labels = []

def extract_features(batch):
    features = {}
    # Iterate over all sensor columns, including bpm
    for col in sensor_columns:
        data = batch[col].values
        # Statistical features
        features[f"{col}_mean"] = np.mean(data)
        features[f"{col}_std"] = np.std(data)
        features[f"{col}_min"] = np.min(data)
        features[f"{col}_max"] = np.max(data)
        features[f"{col}_range"] = np.ptp(data)
        features[f"{col}_iqr"] = iqr(data)
        features[f"{col}_rms"] = np.sqrt(np.mean(np.square(data)))
        features[f"{col}_skew"] = skew(data)
        features[f"{col}_kurtosis"] = kurtosis(data)
        # Time-series features
        features[f"{col}_cumsum"] = np.sum(np.cumsum(data))
        features[f"{col}_zero_crossing"] = np.sum(np.diff(np.sign(data)) != 0)
        features[f"{col}_entropy"] = entropy(np.histogram(data, bins=10)[0] + 1)
        # Frequency domain features (for bpm these might be less relevant, but included for uniformity)
        fft_values = np.abs(fft(data))
        features[f"{col}_fft_peak"] = np.max(fft_values)
        freqs, psd = welch(data, nperseg=min(batch_size, len(data)))
        features[f"{col}_spectral_energy"] = np.sum(psd)
        features[f"{col}_dominant_freq"] = freqs[np.argmax(psd)] if len(freqs) > 0 else 0

    # Combined magnitude features for accelerometer and gyroscope (exclude bpm)
    acc_mag = np.sqrt(batch["AcX"]**2 + batch["AcY"]**2 + batch["AcZ"]**2)
    gyro_mag = np.sqrt(batch["GyX"]**2 + batch["GyY"]**2 + batch["GyZ"]**2)
    features["acc_mag_mean"] = np.mean(acc_mag)
    features["acc_mag_std"] = np.std(acc_mag)
    features["gyro_mag_mean"] = np.mean(gyro_mag)
    features["gyro_mag_std"] = np.std(gyro_mag)

    return features

for i in range(num_batches):
    batch = df.iloc[i * batch_size : (i + 1) * batch_size]
    features_list.append(extract_features(batch))
    labels.append(batch['label'].mode()[0])  # Batch label as mode

features_df = pd.DataFrame(features_list)
labels = np.array(labels)

print("Extracted features shape:", features_df.shape)
print("Labels shape:", labels.shape)

# === Compute Point-Biserial Correlations (for analysis) ===
correlations = []
for col in features_df.columns:
    corr, _ = pointbiserialr(features_df[col], labels)
    correlations.append((col, corr))

print("Point-Biserial Correlations:")
for feature, corr in sorted(correlations, key=lambda x: abs(x[1]), reverse=True):
    print(f"{feature}: {corr}")

# === Select features based on correlation threshold ===
threshold = 0.4
selected_features = [feature for feature, corr in correlations if abs(corr) >= threshold]
print("\nSelected features (|corr| >= 0.4):", selected_features)

# --- Optional: Visualize correlations ---
selected_corrs = [(f, c) for f, c in correlations if f in selected_features]
selected_corrs = sorted(selected_corrs, key=lambda x: abs(x[1]), reverse=True)
sel_features, sel_corr_values = zip(*selected_corrs)
plt.figure(figsize=(10, 6))
sns.barplot(x=list(sel_corr_values), y=list(sel_features), orient="h")
plt.title("Selected Features with |Point-Biserial Correlation| >= 0.4")
plt.xlabel("Point-Biserial Correlation")
plt.ylabel("Feature")
plt.show()

# === Use only selected features for the model ===
features_df_selected = features_df[selected_features]

# === Normalize Data using the selected features ===
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features_df_selected)
joblib.dump(scaler, "scaler.pkl")  # Save scaler for future use

# === Split into training and testing sets ===
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, labels, test_size=0.2, stratify=labels, random_state=42
)

# === Compute class weights to handle imbalance ===
weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weights = dict(enumerate(weights))
print("Class weights:", class_weights)

# === Build an Improved ANN Model ===
input_dim = X_train.shape[1]
model = models.Sequential([
    layers.Dense(256, input_dim=input_dim),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.35),

    layers.Dense(128),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.35),

    layers.Dense(128),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.35),

    layers.Dense(64),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.35),

    layers.Dense(32),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(16),
    layers.LeakyReLU(alpha=0.01),

    layers.Dense(1, activation='sigmoid')  # Binary classification
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# === Train the model with Early Stopping ===
early_stopping = callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True)

history = model.fit(X_train, y_train, epochs=100, batch_size=42,
                    validation_data=(X_test, y_test), class_weight=class_weights,
                    callbacks=[early_stopping])

# === Evaluate the model using default threshold 0.5 ===
loss, acc_default = model.evaluate(X_test, y_test)
print("Test Accuracy (default threshold 0.5):", acc_default)

# === Evaluate with a custom decision threshold ===
predicted_probs = model.predict(X_test)
custom_threshold = 0.5  # Lower decision threshold
predicted_labels_custom = (predicted_probs > custom_threshold).astype(int).flatten()
acc_custom = accuracy_score(y_test, predicted_labels_custom)
print("Test Accuracy (custom threshold 0.5:", acc_custom)
print("Classification Report (custom threshold 0.5):")
print(classification_report(y_test, predicted_labels_custom))
print("Confusion Matrix (custom threshold 0.5):")
print(confusion_matrix(y_test, predicted_labels_custom))

# === Save the improved ANN model ===
model.save("ann_model_improved.keras")
print("Model saved as 'ann_model_improved.keras'")
```
